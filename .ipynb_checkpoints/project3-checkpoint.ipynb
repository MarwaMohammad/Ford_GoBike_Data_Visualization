```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
df.info()
```

```python
clean_df = df.copy()
```

```python
clean_df.head()
```

## Removing Nan Values from dataframe

```python
clean_df.dropna(inplace=True)
```

## Adding Month, day, weekday, time, date, hour, member_age Columns

```python
clean_df['day'] = pd.DatetimeIndex(clean_df['start_time']).day
```

```python
clean_df['week_day'] = pd.DatetimeIndex(clean_df['start_time']).weekday
```

```python
clean_df['time'] = pd.DatetimeIndex(clean_df['start_time']).time
```

```python
clean_df['date'] = pd.DatetimeIndex(clean_df['start_time']).date
clean_df['hour'] = pd.DatetimeIndex(clean_df['start_time']).hour
clean_df['member_age'] = 2019-clean_df['member_birth_year']
```

```python
clean_df.info()
```

## Fixing the datatype

```python
clean_df.start_time = clean_df.start_time.astype('datetime64')
clean_df.end_time = clean_df.end_time.astype('datetime64')
clean_df.date = clean_df.date.astype('datetime64')
clean_df.time = clean_df.time.astype('str')
clean_df.hour = clean_df.hour.astype('int')

clean_df.bike_id = clean_df.bike_id.astype('object')
clean_df.member_birth_year = clean_df.member_birth_year.astype('int')
```

```python
clean_df.info()
```

### Fixing the time column items

```python
clean_df['time'] = clean_df['time'].apply(lambda x: x[:-7] if '.' in x else x)
```

## Removing not required Columns

```python
elpor_df = clean_df.drop(columns=['end_station_id', 'start_station_id', 'start_station_latitude', 'start_station_longitude', 'end_station_latitude','end_station_longitude'])
```

```python
elpor_df.info()
```

## Exploration

```python
elpor_df.head()
```

## Univariate visualization

### What is the Distribution of Age?

```python
sns.histplot(data=elpor_df, x="member_age");
plt.title("Age Distribution", fontsize=11);
```

The distribution of age is right-skewed. Most of people who use the bikeshare service are from 20 to 45.

### What is the Distribution of Rides Count on Days?

```python
sns.histplot(data=elpor_df, x="day", binwidth=2);
plt.title("Rides Distribution on Days", fontsize=11);
```

Here from above figure, we notice that the most two sequential days for sharing bikes is on 19,20 

### What is the Distribution of  rides per Weekday?

```python
sns.histplot(data=elpor_df, x="week_day");
plt.title("Bikeshare state by weekday", fontsize=11);
```

From above figure, we can say that the day 3 has the most rideshares.

### What is the  Percentage for each Gender  in Bikesharing System?

```python
data = elpor_df.member_gender.value_counts()
plt.pie(x=data.values, autopct="%.1f%%", labels=data.index.values, pctdistance=0.5)
plt.title("Bikeshare by Gender", fontsize=11);
```

From above figure, we see that the male is the most gender percentage.

### What is the  Percentage for each User Type  in Bikesharing System?

```python
usertype_data = elpor_df.user_type.value_counts()
plt.pie(x=usertype_data.values, autopct="%.1f%%", labels=usertype_data.index.values, pctdistance=0.5)
plt.title("Bikeshare by User Type", fontsize=11);
```

From above, we can say that the user type subscriber has the most share with us.

### What the Name of the most Station has Bikesharing System?

```python
station_count = elpor_df.start_station_name.value_counts()[:10]
```

```python
color = sns.color_palette()[0]
sns.barplot(x=station_count.values, y=station_count.index.values, color=color)
plt.title("The most common station", fontsize=11);
plt.xlabel('Count');
```

### What is the Share State per each hour in day?

```python
hour_count = elpor_df.hour.value_counts()
```

```python
color = sns.color_palette()[0]
plt.figure(figsize=(10, 8))
sns.barplot(y=hour_count.values, x=hour_count.index.values, color=color)
plt.xticks(rotation=30);
plt.title("Daily share state", fontsize=11);
plt.ylabel('Count');
```

>Here, we notice the followings:
><li> The number of shares has largest between 8 to 10 and 4pm to 7pm. </li>
><li> The sharing bike has contant usage between 7 am to 8 pm. </li>

## Bivariate Visualization

### Which Generation use Bikeshare longer?

```python
sns.relplot(data = elpor_df, x='member_birth_year', y='duration_sec' )
plt.title("Which Generation use Bikeshare longer", fontsize=11);
plt.ylabel('Duration (Sec)');
```

```python
sns.relplot(data = elpor_df, x='member_age', y='duration_sec' )
plt.title(" Which Ages use Bikeshare longer", fontsize=11);
plt.ylabel('Duration (Sec)');
```

Here in the above figure, we notice that people who born after 1960, or less than 60 who share bikes more than others.

### Which Member Gender us Bikesharing longer?

```python
color = sns.color_palette()[0]
plt.figure(figsize=(4, 5))

sns.barplot(data = elpor_df, y='duration_sec', x='member_gender', color=color)
plt.title("Which Member gender us bikesharing longer", fontsize=11);
plt.ylabel('Duration (Sec)');
```

From above figure, we notice that usually the other gender use the bike longer than male or felmale.

### Which User Type share Bikes longer?

```python
color = sns.color_palette()[0]
plt.figure(figsize=(4, 5))

sns.barplot(data = elpor_df, y='duration_sec', x='user_type', color=color)
plt.title("which user type share bikes longer", fontsize=11);
plt.ylabel('Duration (Sec)');
```

From above figure, we notice that the Customer user type use the bike longer than male or Subscriber.

### How many seconds used per each ride on each weekday?

```python
color = sns.color_palette()[0]
sns.barplot(data = elpor_df, y='duration_sec', x='week_day', color=color)
plt.title("How many seconds used per each weekday", fontsize=11);
plt.ylabel('Duration (Sec)');
```

From above figure, The day 6 has the longest duration.

### How many seconds used per each hour on average?

```python
sns.relplot(data = elpor_df, y='duration_sec', x='hour', kind='line')
plt.title("How many seconds used per each hour on average", fontsize=11);
plt.ylabel('Duration (Sec)');
```

From the above graph, we notice the the longest average duration is between 1 am to 4 am but all trips which on avaverage about 10 minutes or 600 seconds will be between 4 am to 11 pm

## Mulitivariate Visualization

### What the Trends of each Gender Type in the Relation between Duration_sec and Hour?

```python
sns.relplot(data = elpor_df, y='duration_sec', x='hour', col='member_gender')
```

### What the Trends of each User Type in the Relation between Duration_sec and Hour?

```python
sns.relplot(data = elpor_df, y='duration_sec', x='hour', col='user_type');

```

### What is the Share Duration for each Gender?

```python
sns.relplot(data = elpor_df, y='duration_sec', x='hour', kind='line', hue='member_gender')
plt.title( 'Share duration for each gender' , size = 10 ) 
plt.ylabel('Duration (Sec)');
```

### What is the Share Count per Day considering Member_gender?

```python
bikes_per_day = elpor_df.groupby(['week_day','member_gender'])['bike_id'].count()
bikes_per_day = bikes_per_day.reset_index()
```

```python
sns.relplot(data = bikes_per_day, y='bike_id', x='week_day', hue='member_gender', style='member_gender', kind='line', markers=True, dashes=False)
plt.title( 'Share count per day considering member_gender' , size = 10 );
plt.ylabel('Bike Count');
```

### What is the Share Count per Day considering User_type?

```python
bike_per_day = elpor_df.groupby(['week_day','user_type'])['bike_id'].count()
bike_per_day = bike_per_day.reset_index()
bike_per_day
```

```python
sns.relplot(data = bike_per_day, y='bike_id', x='week_day', hue='user_type', style='user_type', kind='line', markers=True, dashes=False)
plt.title( 'Share count per day considering user_type' , size = 10 ) 
plt.ylabel('Bike Count');
```

### What is Bikeshare Count per Day considering User_Type?

```python
bike_per_d = elpor_df.groupby(['day','user_type'])['bike_id'].count()
bike_per_d = bike_per_d.reset_index()
```

```python
sns.relplot(data = bike_per_d, y='bike_id', x='day', hue='user_type', style='user_type', kind='line', markers=True, dashes=False)
plt.title( 'Share count per day considering user_type' , size = 10 ) 
plt.ylabel('Bike Count');
```
